## Clean up

This notebook deletes all the runs and the experiment generated by the the other notebooks in this repository. It also removes of S3 object with prefix `s3://<sagemaker default bucket>/pytorch-demo-mnist>`

---
This notebook was developed using Python 3 (Data Science) kernel in SageMaker Studio

---

```python
%%capture
%pip install -U "sagemaker >= 2.123.0"
```

```python
from sagemaker import Session
from sagemaker.experiments.experiment import _Experiment
```

```python
sagemaker_session = Session()
```

Delete the experiment and all its runs

```python
experiment_name = "pytorch-demo-mnist"
exp = _Experiment.load(experiment_name=experiment_name, sagemaker_session=sagemaker_session)
exp._delete_all(action="--force")
```

Delete the model artifacts and other training outputs

```python
s3 = sagemaker_session.boto_session.client("s3")
```

```python
response = s3.list_objects_v2(
    Bucket=sagemaker_session.default_bucket(),
    Delimiter="string",
    Prefix="pytorch-demo-mnist",
)
response = s3.delete_objects(
    Bucket=sagemaker_session.default_bucket(),
    Delete={
        "Objects": [{"Key": k["Key"]} for k in response["Contents"]],
    },
)
```
